```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (30, 20)
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display, clear_output
from io import StringIO
import time

def env_render(env_frames):
    if isinstance(env_frames[0],StringIO):
        for i in range(len(env_frames)):
            if i>0:
                clear_output()
            print(env_frames[i].getvalue())
            time.sleep(1)
    else:
        fig = plt.figure()
        plt.axis('off')

        plot = plt.imshow(env_frames[0])

        def init():
            pass

        def update(i):
            plot.set_data(env_frames[i])
            return plot,

        anim = FuncAnimation(
            plt.gcf(),
            update,
            frames=len(env_frames),
            init_func=init,
            interval=20,
            repeat=True,
            repeat_delay=20)
        plt.close(anim._fig)
        display(HTML(anim.to_jshtml()))

def env_info(env):
    print(env.spec.id)
    print('Action Space:', env.action_space)
    print('Observation Space:', env.observation_space)
    print('Reward Range:', env.reward_range)
    env.reset()
    print('Step output:',env.step(env.action_space.sample()))
```

## Naive Random Policy

```python
def policy_random(env,obs):
    return env.action_space.sample()

# t_max = 0 : run until done
# t_max > 0 : run each episode for t_max steps at max
# g_max = 0 : run until done
# g_max > 0 : run each episode for g_max return at max
import numpy as np
def experiment(env, policy, n_episodes=1, g_max=0, t_max=0):
    if 'rgb_array' in env.metadata['render.modes']:
        render_mode = 'rgb_array'
    else:
        render_mode = 'ansi'
    
    frames=[None] * n_episodes
    returns=np.empty(shape=(n_episodes))
    
    for i in range(n_episodes):
        obs = env.reset()
        done = False
        f = []
        g = 0
        t = 0
        while not done:
            action = policy(env,obs)
            obs, reward, done, info = env.step(action)
            f.append(env.render(mode = render_mode))
            g += reward
            t += 1
            if g_max > 0 and g >= g_max:
                break
            if t_max > 0 and t >= t_max:
                break
        returns[i]= g
        frames[i] = f
        
    print('Policy:{}, Return(G)- Mean: {} Min:{}, Max:{}'
          .format(policy.__name__,
                  np.mean(returns),
                  np.min(returns),
                  np.max(returns)))

    return returns, frames
```

```python
env = gym.make('CartPole-v0')
returns, frames = experiment(env, policy_random)
env.close()
```

## Simple policies

### logic based and random

```python
def policy_logic(env,obs):
    return 1 if obs[2] > 0 else 0
def policy_random(env,obs):
    return env.action_space.sample()
```

```python
# experiment function without collecting frames

# t_max = 0 : run until done
# t_max > 0 : run each episode for t_max steps at max
# g_max = 0 : run until done
# g_max > 0 : run each episode for g_max return at max
import numpy as np
def experiment(env, policy, n_episodes=1, g_max=0, t_max=0):
    
    returns=np.empty(shape=(n_episodes))
    
    for i in range(n_episodes):
        obs = env.reset()
        done = False
        g = 0
        t = 0
        while not done:
            action = policy(env,obs)
            obs, reward, done, info = env.step(action)
            g += reward
            t += 1
            if g_max > 0 and g >= g_max:
                break
            if t_max > 0 and t >= t_max:
                break
        returns[i]= g
        
    print('Policy:{}, Return(G)- Mean: {} Min:{}, Max:{}'
          .format(policy.__name__,
                  np.mean(returns),
                  np.min(returns),
                  np.max(returns)))

    return returns
```

```python
n_episodes = 100
g_max = 10000
env = gym.make('CartPole-v0')
returns=experiment(env,policy_random, n_episodes, g_max)
returns=experiment(env,policy_logic, n_episodes, g_max)
env.close()
```

### parameterized random

```python
def policy_logic(theta,env,obs):
    return 1 if obs[2] > 0 else 0
def policy_random(theta,env,obs):
    return env.action_space.sample()
#def policy_logic(theta,obs):
#    return 1 if obs[2] > 0 else 0
def policy_theta_random(theta,env,obs):
    return 0 if np.matmul(theta,obs) < 0 else 1
```

```python
# experiment function with parameters

# t_max = 0 : run until done
# t_max > 0 : run each episode for t_max steps at max
# g_max = 0 : run until done
# g_max > 0 : run each episode for g_max return at max
import numpy as np
def experiment(env, policy, n_episodes=1, g_max=0, t_max=0):
    
    returns=np.empty(shape=(n_episodes))
    
    for i in range(n_episodes):
        obs = env.reset()
        done = False
        g = 0
        t = 0
        theta = np.random.rand(4) * 2 - 1
        while not done:
            action = policy(theta,env,obs)
            obs, reward, done, info = env.step(action)
            g += reward
            t += 1
            if g_max > 0 and g > g_max:
                break
            if t_max > 0 and t > t_max:
                break
        returns[i]= g
        
    print('Policy:{}, Return(G)- Mean: {} Min:{}, Max:{}'
          .format(policy.__name__,
                  np.mean(returns),
                  np.min(returns),
                  np.max(returns)))

    return returns
```

```python
n_episodes = 100
g_max = 10000
env = gym.make('CartPole-v0')
returns=experiment(env,policy_theta_random, n_episodes, g_max)
returns=experiment(env,policy_random, n_episodes, g_max)
returns=experiment(env,policy_logic, n_episodes, g_max)
env.close()
```
